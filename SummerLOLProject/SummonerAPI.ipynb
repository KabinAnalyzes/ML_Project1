# Parse Match ID's

```python
#Define Parsing Function
from tqdm.notebook import trange, tqdm
match_count=1
player_routing= 'americas'
match_id_list=[]
##  Parses for 1 match from each PUUID and provides progress bar.
def Match_id_parse (PuuidFile):
     with open(PuuidFile , 'r' ,encoding='utf-8') as f:
            for item in f:
                for i in trange(len(f.readlines())):
                    match_id = lol_watcher.match.matchlist_by_puuid(player_routing,item.strip(),count=match_count, type= 'ranked')
                    match_id_list.append(match_id)
                    time.sleep(1)
# Define Save Function
## Exports saved list to txt file.
def TxtExport(FileName):
    with open(FileName, 'w',encoding="utf-8") as f:
        for item in match_id_list:
            # write each item on a new line
            f.write("%s\n" % item)
        print('Done')
```

```python
Match_id_parse('../SummerLOLProject/ChallengerSummonerPUUID.txt')
Match_id_parse('../SummerLOLProject/GMSummonerPUUID.txt')
Match_id_parse('../SummerLOLProject/MasterSummonerPUUID.txt')
TxtExport('../SummerLOLProject/allmatchIds.txt')
      
```
